# Linear Algebra.. Advanced Matrix Operation

```python
from numpy import linalg
import numpy as np
```

# Solving a set of linear equations 
> 2x + 2y+3z = 5 <br>
> 3x + y +4z = 7 <br>
> 4x + 3y =10

```python
a = np.array([[2,2,3], [3,1,4],[4,3,0]])
b = np.array([5,7,10])
x = np.linalg.solve(a, b)
x
```

# Matrix Inversion

```python
a = np.array([[1, 2], [3, 4]])
np.linalg.inv( a )
```
